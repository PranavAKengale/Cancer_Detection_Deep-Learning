```python
X=df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

# https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw

model.add(Dense(units=30,activation='relu'))

model.add(Dense(units=15,activation='relu'))


model.add(Dense(units=1,activation='sigmoid'))

# For a binary classification problem
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1
          )
```

```python
loses=pd.DataFrame(model.history.history)
```

```python
loses.plot()   ### this indicating overfitting data means we are using too much epochs
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

# https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw

model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.5))


model.add(Dense(units=1,activation='sigmoid'))

# For a binary classification problem
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
predict_x=model.predict(X_test) 
classes=np.argmax(predict_x,axis=1)
```

```python
classes
```

```python
from sklearn.metrics import classification_report,confusion matrix
```

```python
print(confusion_matrix(y_test,classes))
```

```python
print(classification_report(y_test,classes))
```
